```python
#This dictionary states for each strain its decendants.
hierarchy = {"WT":['0-xylE','N-xylA*'],
            '0-xylE':['0-topA'],
            '0-topA':['0-crp','N-crp*'],
            '0-crp':['0-yjiY'],
            '0-yjiY':['1-mlc+2','2-icd'],
            '1-mlc+2':['1-malE'],
            '1-malE':['1-thrA+2'],
            '1-thrA+2':['1-prs+2'],
            '1-prs+2':[],
            '2-icd':['2-fli'],
            '2-fli':['2-mlc'],
            '2-mlc':['2-prs+7'],
            '2-prs+7':['2-cbdA'],
            '2-cbdA':[],
            'N-xylA*':[],
            'N-crp*':[],
            }

#the times used - in this example I simply used the index of the measurements.
times = range(0,len(mutfreq["0-crp"]))

abundances = {}
for node in hierarchy:
    abundances[node]=[]

def set_abundances(node,t):
    sz = 0
    for son in hierarchy[node]:
        son_size = set_abundances(son,t)
        sz += son_size
    my_size = max(mutfreq[node][t],sz)
    abundances[node].append(my_size-sz)
    return my_size

for t in range(len(times)):            
    set_abundances("WT",t)
    

#to make the curves smooth we need much more time points.
softtimes = np.linspace(0,len(abundances["0-crp"]),300)


sizes = {} #sizes stores, for each strain, the size each "slice" of it occupies at every time point.
pointabdc = {}  #pointabdc stores, for each strain and every time point, the beginning and ending vertical coordinates
                # of each of its slices.

#initialize the sizes dictionary
for i in hierarcy:
    dec = len(hierarchy[i])
    ser = []
    for j in abundances[i]:
        ser.append(float(j)/(dec+1))
    sizes[i]=ser
    pointabdc[i] = []

#two recursive functions needed to initialize the pointabdc dictionary
def calc_size(node,t):
    nodesize = abundances[node][t]
    for son in hierarchy[node]:
        nodesize+=calc_size(son,t)
    return nodesize

def calc_splits(node,t,offset):
    slice_size = sizes[node][t] 
    points = [offset,offset+slice_size]
    for son in hierarchy[node]:
        calc_splits(son,t,points[-1])
        sz = calc_size(son,t)
        points.append(points[-1]+sz)
        points.append(points[-1]+slice_size)
    pointabdc[node].append(points)
    
for t in range(len(times)):
    calc_splits("WT",t,0) #for every time point do a recursive calculation of the slices each strain occupies

### generating the plotting data    
colors = {"WT":"#0000ff",
            '0-xylE':"#00ff00",
            '0-topA':"#ff0000",
            '0-crp':"#00ffff",
            '0-yjiY':"ff00ff",
            '1-mlc+2':"#ffff00",
            '1-malE':"#ffffff",
            '1-thrA+2':"#000000",
            '1-prs+2':"#000080",
            '2-icd':"#008000",
            '2-fli':"#800000",
            '2-mlc':"#008080",
            '2-prs+7':"#800080",
            '2-cbdA':"#808000",
            'N-xylA*':"#808080",
            'N-crp*':"#606060"
            }

fig = mpl.figure(figsize = (8,6))
plt = fig.add_subplot(111)

for node in pointabdc:
    for i in range(len(pointabdc[node][0])/2):
        pts = []
        pts2 = []
        for t in times:
            pts.append(pointabdc[node][t][2*i])
            pts2.append(pointabdc[node][t][2*i+1])
            
        softpts = spline(times,pts,softtimes)
        softpts2 = spline(times,pts2,softtimes)
        diffs = softpts2-softpts
        realtimes = []
        realpts = []
        realpts2 = []
        for t in range(len(softtimes)):
            if diffs[t]>0.7: # a threshold to use to avoid the lines of every strain start from the beginning of time and end at the end of time
                realtimes.append(softtimes[t])
                realpts.append(softpts[t])
                realpts2.append(softpts2[t])
        plt.fill_between(realtimes,realpts,realpts2,color=colors[node])
        plt.set_ylim(0,100)
        
mpl.savefig("muller-demo2.pdf")
```
